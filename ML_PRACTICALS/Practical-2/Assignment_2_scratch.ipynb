```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression  
```

```python
df=pd.read_csv("temperatures.csv") 
df.isnull().sum()
```

```python
df.head()
```

```python
model=LinearRegression() 
df.columns
```

```python
y=df[['ANNUAL']]
x=df[['YEAR']] 
model.fit(x,y) 
```

```python
model.coef_
```

```python
model.intercept_
```

```python
predict=model.predict(x) 
predict
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score 
mae=np.mean(abs(x-predict)) 
mse=np.mean((x-predict)**2)
mae
```

```python
print(mean_absolute_error(x,predict)) 
```

```python
mse
```

```python
mean_squared_error(x,predict) 
```

```python
np.sqrt(mse)  
```

```python
plt.title("Linear Regression Visualization") 
plt.xlabel("YEARS") 
plt.ylabel("Annual Years") 
plt.scatter(x,y,color="r") 
plt.plot(x,predict,color="b") 
plt.show()
```

```python
r2_score(y,predict) 
```

```python
model.score(x,y) 
```
