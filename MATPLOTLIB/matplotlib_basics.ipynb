```python
import matplotlib.pyplot as plt 
x=[1,2,3,4] #x-co-ordinates
y=[10,20,15,25]  # corresponding y-co-ordinates 
#x-> x-axis , y-> y-axis as arguments 
plt.plot(x,y) 
plt.show() 
#important terms in matplotlib: 
""" 
  Data Point -> (x,y) co-ordinate (single point) 
  Figure -> its like the region of your graphs (like a paper) 
  Axes -> Area of ploting (the actual graph) 
  Plot -> line/bar/point inside the axes 
  Marker -> Dot that represents the Data Point of the graph 
  Line -> The line drawn through the Data Point's to create a connection 
                between them 
  Line Style -> single,bold,dotted,dashted (it is similar to what i did in the DDA algo
                of computer graphics subject, like the concept) 
  Color -> color for the Data Point/Marker/Line 
  legend -> It tells us what the color and the Lines represents in the graph 
  label -> The alias for the x and y axis of the graph 
  title -> headline of the graph 
  Grid -> horizontal and vertical lining => is done on the graph(Like a graph paper) 
  Function -> code => that achieves some behaviour/tasks on the graph 
  method -> class cha function is method 
  Parameters -> values given to the function (it is used to change the graphical representation) 
  keyword Arguments -> the parameters that we are supposed to  assign some value to (ex. color="red") 
  DPI(Dots per inch) -> resolution of the points/Dots on the graph (high->clear image else distored) 
  Backend -> system used to save the graphs (TkAgg,QtAgg,GTK3Agg) 
"""
```

```python
import matplotlib.pyplot as plt 
x=["Mon","Tues","Wed","Thur","Fri"] 
y=[2,7,15,19,13] 
plt.plot(x,y) # arrays was passed as the parameter 
# Setting the attributes of the graph
plt.title("Bakery Sales This Week!") # title of the graph
plt.xlabel("Day of the week") # x-axis label 
plt.ylabel("Sales of the Day") # y-axis label
plt.show() #<- this is responsible for showing the graph  
```

```python
#plt.pie(values,label=label_list,color=color_list,autopct='%1.1f%%')  
#PIE_CHARTS 
import matplotlib.pyplot as plt
regions=['North','South','East','West'] 
revenue=[3000,2000,1500,1000] 

plt.pie(revenue,labels=regions,autopct='%1.1f%%',colors=['gold','skyblue','lightgreen','coral']) 
plt.title('Revenue contribution by region') 
plt.show()
```

```python
#HISTOGRAMS 
#plt.hist(data,bins=number_of_bins,color='colorname',edgecolor='black') 
import matplotlib.pyplot as plt 
scores=[45,67,89,56,78,88,82,60,97,59,30,45,55,69,91,81] 
plt.hist(scores,bins=8,color="red",edgecolor="black") 
plt.xlabel("Score Range") 
plt.ylabel("Number of Students") 
plt.title("Score Distribution of students") 
plt.show() #Height of bar=No.of students that fall in that range
```

```python
#Scatter plots: 
import matplotlib.pyplot as plt 
#plt.scatter(x,y,color="color_name",marker="marker_style",label="label_name") 
hours_studied=[1,2,3,4,5,6,7,8] 
exam_scores=[55,77,69,75,79,81,89,82] 
plt.scatter(hours_studied,exam_scores,color="blue",marker="x",label="Student Data-Class-A") 
plt.scatter([1,2,3,4,5,6,7,8],[57,78,70,73,74,84,91,97],color="orange",marker="x",label="Student Data-Class-B") 
plt.xlabel("Hours studied") 
plt.ylabel("Exam scores") 
plt.title("Hours vs Exam for two classes") 
plt.legend() 
plt.grid(True) 
plt.show() 
```
