```python
#plt.subplot(nrows,ncols,index(1-based))  
import matplotlib.pyplot as plt 

x=[1,2,3,4] 
y=[10,20,15,25] 
plt.subplot(1,2,1)#1-row,2-column madye  
plt.plot(x,y) 
plt.title('Line chart')
plt.subplot(1,2,2) #1st row,2nd column,2nd subplot 

plt.bar(x,y) 
plt.title("Bar Chart") 
plt.tight_layout() 
plt.show()
```

```python
import matplotlib.pyplot as plt 

x=[1,2,3,4] 
y=[10,20,15,25]  
#fig,ax= plt.subplots(nrows,nccols,figsize=(width,height)) 
fig,ax=plt.subplots(1,2,figsize=(10,5)) 

ax[0].plot(x,y,color="blue") 
ax[0].set_title('Line Plot') 
ax[1].bar(x,y,color="green") 
ax[1].set_title('Bar Chart')  
fig.suptitle("Comparison of Line and Bar Charts")  
plt.tight_layout() #this is used to adjust the subplots
plt.show()
```
