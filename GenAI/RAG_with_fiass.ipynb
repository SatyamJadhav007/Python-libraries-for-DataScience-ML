```python
#Import Libraries
from langchain.docstore.document import Document #wrapping used on raw Data before Embedding
from langchain.vectorstores  import FAISS #Database to store our embeddings
from langchain_huggingface import HuggingFaceEmbeddings,ChatHuggingFace,HuggingFaceEndpoint
from langchain_core.messages import (HumanMessage,SystemMessage) #Used for taking Inputs for LLM
from IPython.display import display,Markdown
```

#Embeddings

```python
#Check how many documents we have (Emeddings are created from the documents)
print(f"We have {len(documents)} documents")
```

```python
docs=[Document(page_content=text) for text in documents] #getting documents as a Document() repersentation for every_text/index element
#i.e. it is going to act as a page =>Every text
```

```python
#Using transformer-based embedding model
embedding_model=HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
#Creating a FAISS vector store for the docs
faiss_store=FAISS.from_documents(docs,embedding_model)
```

```python
index=faiss_store.index
print(f"Total number of indexes:{index.ntotal}")  #Print total number of indexes
print(f"Total number of dimensions:{index.d}") #Each doc text has 384 dimensions!!
print(f"Embeddings for the fist vector:\n{index.reconstruct(0)}") #Embeddings for the fist vector
```

```python
#Indexing => vectors in index
#Searching => check for nearest neighbours in the index
#Returning Results => Similar Items based on the query index
#Flat index => Brute-force way for checking for indexes
#IVF(Inverted File) =>divides the file into clusters
#HNSW =>Graph based aproach for fast searches


#Retrive Relevent documents
query="What is Berlin known for?"
k=5
retrieved_docs=faiss_store.similarity_search_with_score(query,k) #based on distance it is (in vector dimension)!
```

```python
#function for retriving docs
def get_relevant_documents(query,k=5):
  return faiss_store.similarity_search(query,k)
```

#Generative System

```python
#FLOW:create a llm instance => set messages(system and human) for it to get the response=>use display-markdown to show the output
#Load the LLM
llm=HuggingFaceEndpoint(
    repo_id="microsoft/Phi-4",
    task="text-generation",
)
chat_model=ChatHuggingFace(llm=llm)
```

```python
#Defining the System and human messages
query="What is good outside of Berlin?"

def generative_system(context,query):
  messages=[
      SystemMessage(content=f"""You are a english tour guide with a thick German Accent.
       Only answer with information from {context}.
       if you don't have information,say you don't know"""),
      HumanMessage(content=f"answer the {query} based on the {context}")
  ]
  ai_output=chat_model.invoke(messages)
  return display(Markdown(ai_output.content))
```

#Combining the Retrieval and Generative  System

```python
#Build the RAG system
def rag(query):
  context=get_relevant_documents(query)
  return generative_system(query,context)
```

```python
#Test the RAG
query="What happens in Munich?"
rag(query)
```

```python
#Prepare test queries
query_list=[
    "Where is the German Parliament?", #The info on this are present in our docs
    "Who invented the pastel de Nata?" #This is irrelevent to our docs
]
```

```python
#Testing the RAG for multiple queries
for query in query_list:
  rag(query)
#I learned how we can use fiass datastore to compare our query with the embeded data stored there to retrive the most relevent info to the query
#mini-Lm was used to embed each document into a 384 dimension vector to numerically compare data <=> query
```
